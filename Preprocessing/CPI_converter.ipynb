```python
import pandas as pd
import re

# Carica il dataset originale
df = pd.read_csv('C:/Users/andre/Documents/GitHub/DMML/dataset/new_imdb.csv', delimiter=';')

# Funzione per estrarre la valuta e pulire il valore del budget
def process_budget(budget):
    if pd.isna(budget):
        return ('USD', None)  # Assumo USD se non specificato e nessun valore numerico
    # Ricerca per simboli di valuta o caratteri non numerici
    currency = re.findall(r'[^\d.]+', str(budget))
    if '$' in budget:
        currency_symbol = 'USD'
    elif currency:
        currency_symbol = currency[0].strip()
    else:
        currency_symbol = 'USD'
    # Pulisce la colonna budget mantenendo solo numeri e punti
    cleaned_budget = re.sub(r'[^\d.]+', '', str(budget))
    return (currency_symbol, cleaned_budget)

# Applica la funzione a ciascun valore nella colonna budget
df['currency'], df['budget'] = zip(*df['budget'].apply(process_budget))

# Salva il nuovo dataset
df.to_csv('new_imdb_currency.csv', index=False)
```

```python
import pandas as pd

# Carica i dati
new_imdb_currency = pd.read_csv('new_imdb_currency_with_nation.csv')
cpi_filtered = pd.read_csv('CPI_filtered.csv', delimiter=';')

# Funzione per aggiustare il budget basandosi sul CPI
def adjust_budget(row, cpi_data):
    country = row['nation']
    year = row['year']
    # Trova la riga in CPI_filtered che corrisponde alla nazione e all'anno
    cpi_row = cpi_data[(cpi_data['Country'] == country) & (cpi_data['year'] == year)]
    if not cpi_row.empty:
        cpi_value = cpi_row['value'].iloc[0]
        # Aggiusta il budget basato sul CPI
        adjusted_budget = row['budget'] / cpi_value
        return adjusted_budget
    return row['budget']  # Restituisci il budget originale se non c'è corrispondenza

# Applica la funzione per ogni riga nel DataFrame dei film
new_imdb_currency['adjusted_budget'] = new_imdb_currency.apply(adjust_budget, axis=1, cpi_data=cpi_filtered)

# Salva i risultati in un nuovo file CSV
new_imdb_currency.to_csv('new_imdb_currency_adjusted.csv', index=False)
```
