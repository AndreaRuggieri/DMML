```python
import pandas as pd

# Carica i dataset
tmdb_data = pd.read_csv('dataset/TMDB_movie_dataset_v11.csv')
imdb_data = pd.read_csv('dataset/IMDb movies.csv')


# Stampa i nomi delle colonne per verificare
print("TMDB Columns:", tmdb_data.columns)
print("IMDb Columns:", imdb_data.columns)

# Merge dei dati basandoti su imdb_title_id (IMDb) e imdb_id (TMDB)
merged_data = pd.merge(imdb_data, tmdb_data, left_on='imdb_title_id', right_on='imdb_id', how='left', suffixes=('_imdb', '_tmdb'))

# Sostituzione condizionata: Da TMDB a IMDb
for idx, row in merged_data.iterrows():
    if pd.notna(row['budget_tmdb']) and row['budget_tmdb'] != 0:
        imdb_data.at[idx, 'budget'] = row['budget_tmdb']
    if pd.notna(row['revenue']) and row['revenue'] != 0:
        imdb_data.at[idx, 'worlwide_gross_income'] = row['revenue']

# Sostituzione condizionata: Da IMDb a TMDB
for idx, row in merged_data.iterrows():
    if pd.notna(row['budget_imdb']) and row['budget_imdb'] != 0:
        tmdb_data.at[idx, 'budget'] = row['budget_imdb']
    if pd.notna(row['worlwide_gross_income']) and row['worlwide_gross_income'] != 0:
        tmdb_data.at[idx, 'revenue'] = row['worlwide_gross_income']

# Salvataggio dei dataset aggiornati
imdb_data.to_csv('test/updated_imdb_data.csv', index=False)
tmdb_data.to_csv('test/updated_tmdb_data.csv', index=False)
```

```python
import pandas as pd

# Carica il dataset
data = pd.read_csv('CPI_filtered.csv')

# Aggiungi una colonna 'Means' inizializzata a NaN
data['Means'] = pd.NA

# Funzione per calcolare la media annuale basata sulla frequenza
def calculate_annual_mean(group):
    if group['Frequency'].iloc[0] == 'Annual':
        group['Means'] = group['Value']
    elif group['Frequency'].iloc[0] == 'Quarterly':
        group['Means'] = group.groupby(group['TIME'].str[:4])['Value'].transform('mean')
        # Mantieni solo la prima riga di ogni anno per i dati trimestrali
        group = group[group['TIME'].str.endswith('Q1')]
    elif group['Frequency'].iloc[0] == 'Monthly':
        group['Means'] = group.groupby(group['TIME'].str[:4])['Value'].transform('mean')
        group = group[group['TIME'].str.endswith('01')]
    return group

# Applica la funzione a ciascun gruppo di paese
result = data.groupby(['Country', 'Frequency']).apply(calculate_annual_mean)

# Salva il nuovo DataFrame con la colonna 'Means' aggiunta
result.to_csv('CPI_means.csv', index=False)
```

```python
import pandas as pd

# Carica i dati
data = pd.read_csv('correct_new_imdb_currency_with_nation.csv')

# Inizializza un contatore per tracciare le righe con errori
error_count = 0

# Passa in rassegna ogni riga del dataset
for index, row in data.iterrows():
    year_str = str(row['year'])
    length = len(year_str)
    
    # Controlla la lunghezza della stringa year e aggiusta il valore
    if length == 3:
        # Aggiungi uno zero alla fine
        data.at[index, 'year'] = year_str + '0'
    elif length == 1:
        # Aggiungi tre zeri alla fine
        data.at[index, 'year'] = year_str + '000'
    elif length == 4:
        # Se contiene esattamente 4 cifre, vai avanti senza fare nulla
        continue
    else:
        # Se non rientra nei casi previsti, stampa un errore
        print(f"Error: Invalid year format at index {index} - {year_str}")
        error_count += 1

# Stampa il numero totale di errori trovati
print(f"Total errors: {error_count}")

# Salva il DataFrame aggiornato, se necessario
data.to_csv('correct_new_imdb_currency_with_nation.csv', index=False)
```

```python
import pandas as pd

# Carica i dati
movies = pd.read_csv('correct_new_imdb_currency_with_nation.csv')
# cpi_data = pd.read_csv('CPI_filtered.csv')
cpi_data = pd.read_csv('CPI_means.csv')
cpi_usa = pd.read_csv('adjusted_inflation.csv')
cpi_usa['Year'] = cpi_usa['Year'].astype(int)

# Preparazione del DataFrame aggiungendo la nuova colonna
movies['budget_with_CPI'] = 0
failures = 0
usa_2023_cpi = 304.65 # cpi of USA in 2023

# Funzione per recuperare il CPI basato su nazione e anno
def get_cpi(data, country, year):
    cpi = data[(data['Country'] == country) & (data['TIME'].str.startswith(str(year)))]
    if not cpi.empty:
        return cpi.iloc[0]['Means']  # Assume che ci sia una sola riga per nazione e anno
    
    return None

def get_usa_cpi(cpi_data, film_year):
    #cpi = data[data['Year'].str.startswith(str(film_year))]
    cpi = cpi_data[cpi_data['Year'] == film_year]
    if not cpi.empty:
        return cpi.iloc[0]['Mean']  # Assume che ci sia una sola riga per nazione e anno
    return None

# Ciclo attraverso ogni film
for index, row in movies.iterrows():
    nation = row['nation']
    film_year = row['year']
    budget = row['budget']

    # Salta le righe dove la nazione è 'United States'
    if nation == 'United States':
        # Ottieni i valori CPI per l'anno del film e per il 2023
        cpi_film_year = get_usa_cpi(cpi_usa, film_year)
    
        if cpi_film_year is None:
            reason = ""
            if cpi_film_year is None: 
                reason = "CPI FILM YEAR,"
            else:
                reason = "CPI 2023,"
            print(f"{reason} Failed to find CPI data for {nation} in year {film_year} or 2023")
            failures += 1
            movies.drop(index, inplace=True)
            continue
    
        # Calcola il budget aggiustato
        if cpi_film_year > 0:  # Evita la divisione per zero
            adjusted_budget = (budget * usa_2023_cpi) / cpi_film_year
            movies.at[index, 'budget_with_CPI'] = adjusted_budget
    else:
        # Ottieni i valori CPI per l'anno del film e per il 2023
        cpi_film_year = get_cpi(cpi_data, nation, film_year)
        cpi_2023 = get_cpi(cpi_data, nation, 2023)
    
        if cpi_film_year is None or cpi_2023 is None:
            reason = ""
            if cpi_film_year is None: 
                reason = "CPI FILM YEAR,"
            else:
                reason = "CPI 2023,"
            print(f"{reason} Failed to find CPI data for {nation} in year {film_year} or 2023")
            failures += 1
            movies.drop(index, inplace=True)
            continue
    
        # Calcola il budget aggiustato
        if cpi_film_year > 0:  # Evita la divisione per zero
            adjusted_budget = (budget * cpi_2023) / cpi_film_year
            movies.at[index, 'budget_with_CPI'] = adjusted_budget

# Stampa il numero di righe che non hanno trovato dati CPI
print(f"Total failures: {failures}")

# Salva il nuovo DataFrame
movies.to_csv('budget_with_CPI.csv', index=False)
```

```python
import pandas as pd
from datetime import datetime
 
# Caricare i dataset
df_films = pd.read_csv('revenue_budget_with_CPI.csv')
df_exchange_rates = pd.read_csv('../test/eurofxref-hist.csv')

def add_default_date(date_str):
    if pd.notnull(date_str) and len(date_str) == 4:
        return f"{date_str}-01-01"
    return date_str

df_films['date_published'] = df_films['date_published'].apply(add_default_date)

df_films['date_published'] = pd.to_datetime(df_films['date_published'], format='%Y-%m-%d', errors='coerce')

"""# Funzione per controllare e formattare la data
def format_date(date):
    if len(date) == 4:  # Significa che abbiamo solo l'anno
        return pd.to_datetime(date + '-01-01', format='%Y-%m-%d')
    else:
        return pd.to_datetime(date, format='%Y-%m-%d', errors='coerce')
 
# Convertire le colonne date_published e date in formato datetime
df_films['date_published'] = pd.to_datetime(df_films['date_published'].astype(str) + '-01-01', format='%Y-%m-%d',errors='coerce')"""

df_exchange_rates['Date'] = pd.to_datetime(df_exchange_rates['Date'])
 
# Funzione per trovare la data più vicina
def find_closest_date(date, dates):
    return dates.iloc[(dates - date).abs().argsort()[0]]
 
# Lista per salvare i valori del budget convertito
converted_budgets = []
missing_currencies = set()
error_count = 0
 
# Scorrere ogni film nel dataset
for index, film in df_films.iterrows():
    # Trovare la data più vicina nel dataset dei tassi di conversione
    closest_date = find_closest_date(film['date_published'], df_exchange_rates['Date'])
    exchange_rate_row = df_exchange_rates[df_exchange_rates['Date'] == closest_date]
 
    # Se la valuta è diversa da USD, convertire il budget
    if film['currency'] != 'USD':
        if film['currency'] in exchange_rate_row.columns or film['currency'] == 'EUR':
            if film['currency'] != 'EUR':
                currency_rate = exchange_rate_row[film['currency']].values[0]
            else:
                currency_rate = 1
            eur_rate = exchange_rate_row['USD'].values[0]
            converted_budget = (film['budget_with_CPI'] / currency_rate) * eur_rate
        else:
            missing_currencies.add(film['currency'])
            error_count += 1
            converted_budget = None  # oppure puoi scegliere di inserire il valore originale del budget
    else:
        # Se la valuta è USD, non c'è bisogno di conversione
        converted_budget = film['budget_with_CPI']
 
    # Aggiungere il valore convertito alla lista
    converted_budgets.append(converted_budget)
 
# Aggiungere la nuova colonna al dataset dei film
df_films['converted_budget'] = converted_budgets
 
# Salvare il dataset aggiornato in un nuovo file CSV (opzionale)
df_films.to_csv('updated_films.csv')
```
